```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,1])

plt.show()
```

```python
# adding title and lables

plt.plot([1,2,3],[4,5,1])
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# change line aesthetic

plt.plot([1,2,3],[4,5,1],color='r',linestyle=":",linewidth=3)
plt.show()
```

```python
# adding 2 lines in the same plot

x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'r',linestyle=":",linewidth=3)
plt.plot(x2,y2,'g',linestyle="-",linewidth=3)
plt.title('Epic info')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# adding grid lines to a plot


x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'r',linestyle=":",linewidth=3)
plt.plot(x2,y2,'g',linestyle="-",linewidth=3)
plt.title('Epic info')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

```python
# subplot - display multiple plots in one figure
```

```python
# plot 1:

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

# plot 2:

x = ([0,1,2,3])
y = ([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
# subplot() function : 
# plt.subplot(1,2,1) means the figure has 1 row, 2 columns (for first plot)
# plt.subplot(1,2,2) means the figure has 1 row, 2 columns (for second plot)
```

```python
# changing figure of plot


x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)
plt.plot(x,y,'r',linestyle=":", linewidth=3)
plt.ylabel('Y axis')
plt.xlabel('X axis')

# plot 2:

x = ([0,1,2,3])
y = ([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.plot(x,y,'g',linestyle=":", linewidth=3)
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
# Bar graph : It can be represented horizontly or vertically.
```

```python
categories = ['Category A', 'Category B','Category C','Category D']
values = [20,35,30,25]

plt.bar(categories,values)
plt.show()
```

```python
data = {'Category A':20,'Category B':35,'Category C':30,'Category D':25}

categories = list(data.keys())
values =list(data.values())

plt.xlabel('categories')
plt.ylabel('values')
plt.bar(categories,values)
plt.show()
```

```python
# horizontal bar graph
# use plt.barh (categories,values) 
# ('h' horizontal bar).

categories = ['Category A', 'Category B','Category C','Category D']
values = [20,35,30,25]

plt.barh(categories,values,color='y')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')

plt.show()
```

```python
# comparison or representation of two bar graph 

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label="BMW",width=.3)

plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label="Audi", color='r',width=.3)

plt.legend()

plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')

plt.show()
```
