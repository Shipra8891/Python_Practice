```python
# Pie Charts - display the distribution of the data as slices of a circle.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create the pie chart

fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Mangoes']
sizes = [30, 25, 15, 20, 10]

plt.pie(sizes, labels=fruits)
plt.show()
```

```python
# create heading (plt.title())
# puting vales in % ( autopct='%1.1f%%')

fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Mangoes']
sizes = [30, 25, 15, 20, 10]

plt.pie(sizes, labels=fruits , autopct='%1.1f%%')
plt.title('fruit distribution')
plt.show()
```

```python
# to add list of explanation , use the legend() function

fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Mangoes']
sizes = [30, 25, 15, 20, 10]

plt.pie(sizes, labels=fruits, autopct='%1.1f%%')
plt.title('fruit distribution')

plt.legend()

plt.show()
```
